````python
#Se separa el dataset en 70% train, 20% test y 10% validation
train_data, dev_data = train_test_split(df, test_size=0.3, random_state=42)
validation_data, test_data = train_test_split(dev_data, test_size=0.33, random_state=42)

#Se ven tamaños
print(f'Tamaño train: {train_data.shape[0]}, validation: {validation_data.shape[0]}, test: {test_data.shape[0]}')
````

P1.2

````python
#Se crea la función para extrar el día, mes y año
def extract_date_features(df):
    df['day'] = pd.to_datetime(df['date'], format='%d/%m/%y').dt.day.astype(int)
    df['month'] = pd.to_datetime(df['date'], format='%d/%m/%y').dt.month.astype(int)
    df['year'] = pd.to_datetime(df['date'], format='%d/%m/%y').dt.year.astype(int)
    return df[['day', 'month', 'year']]

#Se implementa FunctionTransformer
date_transformer = FunctionTransformer(extract_date_features)

# Aplicar el transformer
date_features = date_transformer.fit_transform(train_data)
date_features.head()
````

P1.3

````python
#Variables categóricas
categorical_cols = ['city', 'shop', 'brand', 'container', 'capacity']
#Variables numéricas
numerical_cols = ['lat', 'long', 'price']

#Se crea el ColumnTransformer con OneHotEncoder y el transformer para date
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(sparse_output=False), categorical_cols),
        ('date', date_transformer, ['date'])
    ]
).set_output(transform='pandas')

#Printeo para verificar
print(preprocessor)
````

P1.4

````python
#Se crea el pipeline con el DummyRegressor
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', DummyRegressor(strategy='mean'))
])
````

P1.5

````python
#Se entrena el pipeline
pipeline.fit(train_data, train_data['quantity'])

#Se predice sobre los datos de validation y se calcula el MAE
val_predictions = pipeline.predict(validation_data)
mae = mean_absolute_error(validation_data['quantity'], val_predictions)
print(f'MAE para DummyRegressor: {mae}')
````

>El MAE de 13,308.13 indica que, en promedio, el modelo se equivoca por 13,308.13 unidades respecto a los valores reales. Como el `DummyRegressor` predice solo el promedio, este valor representa el error que sirve como referencia mínima para evaluar modelos más complejos. Desde el punto de vista de negocio, implica que la gestión de inventario sea compleja dado que se podría tener tanto sobrestock (zonas de baja demanda) como stock insuficiente de bebidas (zonas con alta demanda) por el valor de la métrica, de esta manera se podrían desencadenar desde mayores costos de almacenamiento a insatisfacción de clientes, respectivamente.

P1.6

````python
#Se modifica el pipeline para usar XGBRegressor
pipeline_xgb = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', XGBRegressor())
])

#Entrenamiento
pipeline_xgb.fit(train_data, train_data['quantity'])

#Predicción sobre validation
val_predictions_xgb = pipeline_xgb.predict(validation_data)

#Se calcula el nuevo MAE
mae_xgb = mean_absolute_error(validation_data['quantity'], val_predictions_xgb)
print(f'MAE para XGBRegressor: {mae_xgb}')
````

>El MAE para el `XGBRegressor` es **2,246.30**, lo que es significativamente menor que el MAE del `DummyRegressor` (13,308.13). Esto indica que el `XGBRegressor` es mejor ya que reduce el error promedio en comparación con el modelo base, lo que sugiere que está capturando mejor las relaciones en los datos del negocio.

    - `max_leaves` buscando valores enteros en el rango (0, 100)

    - `reg_alpha` buscando valores flotantes en el rango (0, 1)
    - `reg_lambda` buscando valores flotantes en el rango (0, 1)

Para ello se pide los siguientes pasos:

P3.1

````python
optuna.logging.set_verbosity(optuna.logging.WARNING)
#Semilla
SEED = 42

def objective(trial):
    #Optimización de hiperparámetros para OneHotEncoder
    min_frequency = trial.suggest_float('min_frequency', 0.0, 1.0)

    #Optimización de hiperparámetros para XGBRegressor
    learning_rate = trial.suggest_float('learning_rate', 0.001, 0.1)
    n_estimators = trial.suggest_int('n_estimators', 50, 1000)
    max_depth = trial.suggest_int('max_depth', 3, 10)
    max_leaves = trial.suggest_int('max_leaves', 0, 100)
    min_child_weight = trial.suggest_int('min_child_weight', 1, 5)
    reg_alpha = trial.suggest_float('reg_alpha', 0.0, 1.0)
    reg_lambda = trial.suggest_float('reg_lambda', 0.0, 1.0)

    #Se crea el preprocesador
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numerical_cols),
            ('cat', OneHotEncoder(sparse_output=False, min_frequency=min_frequency), categorical_cols),
            ('date', date_transformer, ['date'])
        ]
    ).set_output(transform='pandas')

    #Se crea el pipeline
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', XGBRegressor(
            learning_rate=learning_rate,
            n_estimators=n_estimators,
            max_depth=max_depth,
            max_leaves=max_leaves,
            min_child_weight=min_child_weight,
            reg_alpha=reg_alpha,
            reg_lambda=reg_lambda,
            random_state=SEED,
        ))
    ])

    #Entrenamiento
    pipeline.fit(train_data, train_data['quantity'])

    #Predicción en el conjunto de validación y se calcula el MAE
    val_predictions = pipeline.predict(validation_data)
    mae = mean_absolute_error(validation_data['quantity'], val_predictions)

    #Se guarda el mejor pipeline
    trial.set_user_attr('best_pipeline', pipeline)

    return mae
````

P3.2 - P3.3

````python
#Configuración de Optuna con TPESampler y un tiempo límite de 5 minutos
study1 = optuna.create_study(direction='minimize', sampler=TPESampler(seed=SEED))
study1.optimize(objective, timeout=300)
````

````python
#Printeo de resultados
print(f"Number of trials: {len(study1.trials)}")
print(f"Best MAE: {study1.best_value}")
print(f"Best hyperparameters: {study1.best_params}")

#Se guarda el mejor modelo
best_pipeline = study1.best_trial.user_attrs['best_pipeline']
````

>
Se obtiene un MAE de **1897.41**, menor que el mejor caso anterior por **348.89** unidades aproximadamente. La mejora se debe a la optimización bayesiana de los hiperparámetros, que ajustó valores clave como `learning_rate`, `n_estimators`, y `max_depth` para encontrar una configuración más precisa que el modelo base. Esta metodología explora el espacio de soluciones de manera más eficiente que el ajuste manual o aleatorio, ya que realiza una búsqueda en espacios con mayor probabilidad de reducir el error.
>
>### Explicación de los hiperparámetros optimizados
>
>- **`min_frequency`**: Controla la frecuencia mínima de las categorías en el `OneHotEncoder`. Un valor bajo (0.05) asegura que más categorías sean codificadas.
>- **`learning_rate`**: Controla la velocidad de aprendizaje; un valor moderado (0.06) balancea precisión y convergencia.
>- **`n_estimators`**: Aumentar a 917 mejora la capacidad del modelo para aprender patrones complejos.
>- **`max_depth`**: Mayor profundidad (9) permite que los árboles capturen interacciones más profundas.
>- **`max_leaves`**: Un valor alto (93) aumenta la flexibilidad del modelo sin llegar a sobreajustar.
>- **`min_child_weight`**: Un valor de 4 reduce el sobreajuste en hojas pequeñas.
>- **`reg_alpha` y `reg_lambda`**: Ambos valores (0.67 y 0.91) controlan el sobreajuste, regularizando el modelo.
>
>Respecto a los rangos, estos son adecuados y son empleados en la práctica común  ya que permiten una búsqueda eficiente de configuraciones que optimizan tanto la capacidad de generalización como la capacidad de ajuste del modelo.

P4.1

**Respuestas:**
>Prunning (poda) es una técnica utilizada en la optimización de modelos para detener los trials que parecen ser poco prometedores en fases tempranas del entrenamiento, así evita gastar tiempo en modelos que no proporcionarán buenos resultados. En lugar de completar todos los entrenamientos de los hiperparámetros seleccionados, el pruning interrumpe aquellos que claramente no alcanzarán un buen desempeño, por lo tanto, acelera el proceso de optimización, liberando recursos para probar configuraciones más prometedoras. De esta manera, permite explorar más configuraciones en el tiempo disponible.



P4.2

````python
optuna.logging.set_verbosity(optuna.logging.WARNING)

def objective(trial):
    # Optimización de hiperparámetros para OneHotEncoder y XGBRegressor
    min_frequency = trial.suggest_float('min_frequency', 0.0, 1.0)
    learning_rate = trial.suggest_float('learning_rate', 0.001, 0.1)
    n_estimators = trial.suggest_int('n_estimators', 50, 1000)
    max_depth = trial.suggest_int('max_depth', 3, 10)
    max_leaves = trial.suggest_int('max_leaves', 0, 100)
    min_child_weight = trial.suggest_int('min_child_weight', 1, 5)
    reg_alpha = trial.suggest_float('reg_alpha', 0.0, 1.0)
    reg_lambda = trial.suggest_float('reg_lambda', 0.0, 1.0)

    #Se crea el preprocesador
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numerical_cols),
            ('cat', OneHotEncoder(sparse_output=False, min_frequency=min_frequency), categorical_cols),
            ('date', date_transformer, ['date'])
        ]
    ).set_output(transform='pandas')

    #Se aplica el preprocesador
    X_train_trans = preprocessor.fit_transform(train_data.drop(columns=['quantity']))
    X_val_trans = preprocessor.transform(validation_data.drop(columns=['quantity']))

    #Se crea la DMatrix para xgboost
    dtrain = xgb.DMatrix(X_train_trans, label=train_data['quantity'])
    dval = xgb.DMatrix(X_val_trans, label=validation_data['quantity'])

    #Se definen los hiperparámetros
    params = {
        'objective': 'reg:squarederror',
        'learning_rate': learning_rate,
        'max_depth': max_depth,
        'max_leaves': max_leaves,
        'min_child_weight': min_child_weight,
        'reg_alpha': reg_alpha,
        'reg_lambda': reg_lambda,
        'eval_metric': 'mae',
        'verbosity': 0
    }

    #Conjunto de evaluación para early stopping
    evals = [(dtrain, 'train'), (dval, 'eval')]

    #Habilitar pruning
    pruning_callback = XGBoostPruningCallback(trial, "eval-mae")

    #Se entrena el modelo con early stopping y pruning
    model = xgb.train(
        params,
        dtrain,
        num_boost_round=n_estimators,
        evals=evals,
        early_stopping_rounds=10,
        callbacks=[pruning_callback]
    )

    #Se predice en el conjunto de validación y calcula el MAE
    val_predictions = model.predict(dval)
    mae = mean_absolute_error(validation_data['quantity'], val_predictions)

    #Se guarda el mejor modelo
    trial.set_user_attr('best_model', model)

    return mae

````

````python
#Configuración de Optuna con TPESampler, Prunning y un tiempo límite de 5 minutos
study = optuna.create_study(direction='minimize', sampler=TPESampler(seed=SEED))
study.optimize(objective, timeout=300, show_progress_bar=True)
````
